```python
sns.catplot(data=admission_df, kind="bar",x="facility", y="2021",height=8, aspect=1.5)
plt.title("Number of Admissions to Public Healthcare Facilities in 2021")
plt.xlabel("Facility")
plt.ylabel("Number of Admissions")
plt.xticks(rotation=45)
plt.show()
```

```python
tmp = pd.melt(admission_df[(admission_df.facility!='Polyclinics') & (admission_df.facility!='Specialist Outpatient Clinics') ], id_vars=['facility'],value_vars=['2012','2013','2014','2015','2016','2017','2018','2019','2020','2021'])
tmp.columns=['facility','year','admission']
plt.figure(figsize=(20,15))
sns.lineplot(data=tmp,x='year',y='admission',hue='facility')
```

```python
tmp = pd.melt(admission_df, id_vars=['facility'],value_vars=['2012','2013','2014','2015','2016','2017','2018','2019','2020','2021'])
tmp.columns=['facility','year','admission']
tmp['admission']=tmp['admission'].astype(int)
tmp['facility'].unique()
```

```python
output_notebook()
p = figure(width=1000, height=550)

for name, color in zip(['Acute Hospitals', 'Psychiatric Hospitals', 'Community Hospitals',
       'Specialist Outpatient Clinics',
       'Accident & Emergency Departments', 'Polyclinics',
       'Dental Clinics', 'Day Surgeries'], Category10[8]):
    df = tmp[tmp.facility==name]
    p.line(df['year'], df['admission'], line_width=2, color=color, alpha=0.8, legend_label=name)

p.legend.location = "top_left"
p.legend.click_policy="hide"

show(p)
```

```python
facility = pd.read_csv("otot3_facility.csv")
facility.head()
```

```python
facility_df = facility[(facility['facility']!='  Public')&(facility['facility']!='  Private')&(facility['facility']!='  Not-For-Profit')&(facility['facility']!='Inpatient Hospice Palliative Care Service (IHPCS)')].reset_index(drop=True)
facility_df = facility_df.iloc[:,:11]
facility_df
```

```python
sns.catplot(data=facility_df, kind="bar",x="facility", y="2021",height=8, aspect=1.5)
plt.xticks(rotation=45)
plt.title("Number of Healthcare Facilities in 2021")
plt.xlabel("Facility")
plt.ylabel("Number of Facility")
plt.show()
```

```python
tmp = pd.melt(facility_df, id_vars=['facility'],value_vars=['2012','2013','2014','2015','2016','2017','2018','2019','2020','2021'])
tmp.columns=['facility','year','supply']
tmp['supply']=tmp['supply'].astype(int)
plt.figure(figsize=(20,15))
sns.lineplot(data=tmp,x='year',y='supply',hue='facility')
plt.title("Number of Healthcare Facilities during 2012-2021")
plt.xlabel("Facility")
plt.ylabel("Number of Facility")
```

```python
facility_df.loc[facility_df.facility=='Public - Polyclinics','facility']='Polyclinics'
```

```python
facility_df
```

```python
join_df = pd.merge(admission_df,facility_df,how='inner',on='facility',suffixes=('_admission', '_supply'))
```

```python
join_df = join_df.astype({"2019_supply":"int","2018_supply":"int","2017_supply":"int","2016_supply":"int","2015_supply":"int","2014_supply":"int","2013_supply":"int","2012_supply":"int"})
```

```python
for year in ['2012','2013','2014','2015','2016','2017','2018','2019','2020']:
    for tag in ['admission','supply']:
        join_df[year+'_' + tag +'_YOY'] = (join_df[str(int(year)+1)+'_'+tag]-join_df[year+"_"+tag])/join_df[year+"_"+tag]
```

```python
join_df
```

```python
dfs = []
for year in range(2012, 2021):
    new_cols_name = ["admission_YOY", "supply_YOY"]
    cols = [f"{year}_{col}" for col in new_cols_name]
    tmp_df = join_df[["facility"] + cols].rename(columns=dict(zip(cols, new_cols_name)))
    tmp_df["year"] = year
    dfs.append(tmp_df)
concat_df = pd.concat(dfs, ignore_index=True)
concat_df
```

```python
plt.figure(figsize=(20,15))
sns.scatterplot(data=concat_df, x="admission_YOY", y="supply_YOY", hue="facility",s=200)
plt.title("YOY Growth of Admissions and Facilities during 2012-2021")
plt.xlabel("YOY Growth of Admission")
plt.ylabel("YOY Growth of Facility")
```
